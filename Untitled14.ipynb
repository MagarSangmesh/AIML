```python
import numpy as np
import time
import cv2
import math

labelsPath = '.\coco.names'
LABELS = open(labelsPath).read().strip().split("\n")
print(type(LABELS))

weightsPath = './yolov3.weights'
configPath = './yolov3.cfg'
cap = cv2.VideoCapture('./test_video.mp4')

hasFrame, frame = cap.read()
print(hasFrame)

print('Loading pretrained model')
network = cv2.dnn.readNetFromDarknet(configPath, weightsPath)

vid_writer = cv2.VideoWriter('output.avi', cv2.VideoWriter_fourcc('M','J','P','G'),
10, (frame.shape[1],frame.shape[0] ))

while True:
    ret, img = cap.read()
    if not ret:
        break
    
    img = cv2.resize(img, (640,360))
    print(img.shape)
    (H, W) = img.shape[:2]
    
    layers = network.getLayerNames()
    layers = [layers[i-1] for i in network.getUnconnectedOutLayers()]

    blob = cv2.dnn.blobFromImage(img, 1/255.0, (416,416), swapRB=True, crop=False)

    network.setInput(blob)

    start = time.time()

    layerOutputs = network.forward(layers)

    end = time.time()

    print('Frame Prediction Time: {:.3f} seconds'.format(end-start))

    boxes = []
    confidences = []
    classIDs = []

    for output in layerOutputs:
        for detection in output:
            scores = detection[5:]
            classID = np.argmax(scores)
            confidence = scores[classID]
            if confidence > 0.4 and classID == 0:
                box = detection[0:4] * np.array([W, H, W, H])
                (centroid_x,centroid_y,w,h) = box.astype(int)
                x = int(centroid_x-w/2)
                y = int(centroid_y-h/2)
                boxes.append([x,y,int(w),int(y)])
                confidences.append(float(confidence))
                classIDs.append(classID)
    
    idxs = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.3)
    #ind = []

    #for i in range(0, len(classIDs)):
     #   if(classIDs[i]==0):
      #      ind.append(i)
    
    a = []
    b = []

    if(len(idxs) > 0):
        for i in idxs.flatten():
            (x,y) = (boxes[i][0], boxes[i][1])
            (w,h) = (boxes[i][2], boxes[i][3])
            a.append(x)
            b.append(y)
    
    distance = []
    nsd = []

    for i in range(0, len(a)-1):
        for k in range(1, len(a)):
            if(k==i):
                break
            else:
                x_dist = a[k] - a[i]
                y_dist = b[k] - b[i]
                d = math.sqrt(x_dist * x_dist + y_dist * y_dist)
                distance.append(d)
                if(d <= 100):
                    nsd.append(i)
                else:
                    nsd.append(k)
                nsd = list(dict.fromkeys(tuple(nsd)))

    color = (0,0,255)
    for i in nsd:
        (x,y) = (boxes[i][0], boxes[i][1])
        (w,h) = (boxes[i][2], boxes[i][3])
        cv2.rectangle(img, (x,y), (x+w,y+h), color, 2)
        text='Alert'
        cv2.putText(img, text, (x,y-5), cv2.FONT_HERSHEY_PLAIN, 0.5, color, 2)
    
    color = (0,255,0)
    if(len(idxs)>0):
        for i in idxs.flatten():
            if(i in nsd):
                break
            else:
                (x,y) = (boxes[i][0], boxes[i][1])
                (w,h) = (boxes[i][2], boxes[i][3])
                cv2.rectangle(img, (x, y), (x + w, y + h), color, 2)
                text = 'OK'
                cv2.putText(img, text,(x,y-5), cv2.FONT_HERSHEY_PLAIN, 0.5, color,2)
    
    cv2.imshow("Social Distancing",img)

    if(cv2.waitKey(1) & 0xFF == ord('q')):
        break
    vid_writer.write(img)

vid_writer.release()
cap.release()
cv2.destroyAllWindows()
```
